```python
## datos de tabla ya existente en el servidor 
import pandas as pd
from sqlalchemy import create_engine, text

# Establecer la conexión a la base de datos PostgreSQL
engine = create_engine('postgresql://postgres:postgres@3.69.128.37:5432/postgres')

# Crea un objeto de la clase Connection
conn = engine.connect()

## leer la tabla de hechos distribución 

distribucion = text('SELECT * FROM attsf.tbl_distribucion ORDER BY salida_fecha_hora DESC LIMIT 1')

df_distribucion = pd.read_sql_query(distribucion, con=conn)

## obtener el id_distribucion del ultimo registro más resiente 

ultimo_id = text('SELECT id_distribucion FROM attsf.tbl_distribucion ORDER BY salida_fecha_hora DESC LIMIT 1')

ultimo_id = pd.read_sql_query(ultimo_id, con=conn)


conn.close()
engine.dispose()



## lectura de nuevos datos 

## crear path de donde se almacenan datos 
path_input_nuevos = dependencies.rootFolder / 'data' / 'nuevos_datos'

## leer datos nuevos
df_distribucion_nuevos = pd.read_excel(
    path_input_nuevos / '2024 Datos Distribucion Prueba.xlsx', 
)
df_distribucion_nuevos = df_distribucion_nuevos.drop('id_distribucion', axis=1)
df_distribucion_nuevos.index += (ultimo_id.loc[0, 'id_distribucion'] + 1)
df_distribucion_nuevos = df_distribucion_nuevos.reset_index(names='id_distribucion')
df_distribucion_nuevos
```

```python
from sqlalchemy import create_engine

# Crear una conexión a la base de datos
engine = create_engine('postgresql://postgres:postgres@3.69.128.37:5432/postgres')

# Insertar los nuevos datos en la tabla de la base de datos
df_distribucion_nuevos.to_sql(name='tbl_distribucion', con=engine, schema='attsf', if_exists='append', index=False)

# Cerrar la conexión a la base de datos
engine.dispose()
```
