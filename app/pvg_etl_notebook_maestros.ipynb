## MAESTRO TIPO DE OT
Tabla con los tipos de ot.

```python
data_tipo_ot = ['Correctivo', 'Preventivo']
df_tipo_ot = pd.DataFrame(data_tipo_ot, columns=['tipo_ot'])

#Columna de indices
df_tipo_ot.index += 1
df_tipo_ot.reset_index(names='id_tipo_ot', inplace=True)

df_tipo_ot
```

## MAESTRO FRECUENCIA
Tabla con la frecuencia del mantenimiento.

```python
data_frecuencia = ['No frecuencia', 'Anual', 'Bienal',  'Semestral', 'Cuatrimestral', 'Mensual', 'Semanal']
df_frecuencia = pd.DataFrame(data_frecuencia, columns=['frecuencia'])

#Columna de indices
df_frecuencia.index += 1
df_frecuencia.reset_index(names='id_frecuencia', inplace=True)

df_frecuencia
```

## MAESTRO TIPO TALLER
Tabla con el tipo de taller.

```python
data_tipo_taller = ['TR',  'ATC']
df_tipo_taller = pd.DataFrame(data_tipo_taller, columns=['tipo_taller'])

df_tipo_taller['descripcion']=['Taller Regional', 'Taller Central']

#Columna de indices
df_tipo_taller.index += 1
df_tipo_taller.reset_index(names='id_tipo_taller', inplace=True)

df_tipo_taller
```

## MAESTRO TALLER
Tabla con id_taller, taller, id_tipo_taller e id_wilaya

```python
#Se toman los datos del csv taller processed
df_taller = pd.read_csv(path_excel / 'taller.csv',
                skiprows=1,
                names=['numero', 'tipo_taller', 'taller']
                )

#Se toman las columnas que se quieren
df_taller = df_taller[['taller', 'tipo_taller']]

#Se limpian los datos
df_taller['taller'] = df_taller['taller'].str.lstrip()
df_taller['taller'] = df_taller['taller'].str.rstrip()
df_taller['taller'] = df_taller['taller'].str.title()

df_taller['tipo_taller'] = df_taller['tipo_taller'].str.lstrip()
df_taller['tipo_taller'] = df_taller['tipo_taller'].str.rstrip()
df_taller['tipo_taller'] = df_taller['tipo_taller'].str.upper()

#Se quita Instituciones
df_taller = df_taller.drop(5)

#Merge de taller y tipo_taller
df_taller = pd.merge(df_taller, df_tipo_taller, how='left', on='tipo_taller')
#Eliminar la columna ot
df_taller = df_taller.drop('tipo_taller', axis=1)

#Columna id_wilaya: se meten a mano ya que hay dos talleres que pertenecen a Rabuni (6)
df_taller['id_wilaya'] = [1, 5, 2, 4, 3, 6]

#Reordenar dataframe
df_taller = pd.DataFrame({'id_wilaya': df_taller['id_wilaya'], 'id_tipo_taller': df_taller['id_tipo_taller'], 
'taller': df_taller['taller']})

# Crear la nueva fila del nuevo taller
fila_taller = {'id_wilaya': '6', 'id_tipo_taller': '2', 'taller': 'BDT'}

# Añadir la fila
df_taller = df_taller.append(fila_taller, ignore_index=True)

#Columna de indices
df_taller.index += 1
df_taller.reset_index(names='id_taller', inplace=True)

df_taller
```

## MAESTRO AVERIA
Tabla con las posibles averias.

```python
data_averia = ['Chasis', 'Carroceria', 'Ruedas', 'Mecanica', 'Elec, vehiculos', 'Obra civil', 'Agua y combustible', 'Herramientas',
'Informatica', 'Exteriores', 'Aire', 'Maquinaria', 'Electricidad']
df_averia = pd.DataFrame(data_averia, columns=['averia'])

#Columna de indices
df_averia.index += 1
df_averia.reset_index(names='id_averia', inplace=True)

df_averia
```

## MAESTRO REPUESTO
Tabla con id_repuesto y el repuesto

```python
#Se toman los datos del csv averia-ot processed
df_repuesto = pd.read_csv(path_excel / 'repuesto.csv',
                header=None,
                skiprows=1,
                names=['numero','repuesto']
                )

#Se toman las columnas que se quieren
df_repuesto = df_repuesto[['repuesto']]

#Se limpian los datos
df_repuesto['repuesto'] = df_repuesto['repuesto'].str.lstrip()
df_repuesto['repuesto'] = df_repuesto['repuesto'].str.rstrip()
df_repuesto['repuesto'] = df_repuesto['repuesto'].str.capitalize()

#Columna de indices
df_repuesto.index += 1
df_repuesto.reset_index(names='id_repuesto', inplace=True)

df_repuesto
```

# MAESTRO MECANICOS
Tabla con id_mecanico, el nombre del mecanico y puede que la wilaya asociada.

```python
#data_mecanicos = [{'mecanico': 'No mecanico', 'id_taller': '6'}]
data_mecanicos = []
df_mecanico = pd.DataFrame(data_mecanicos, columns=['mecanico', 'id_taller'])

#Columna de indices
df_mecanico.index += 1
df_mecanico.reset_index(names='id_mecanico', inplace=True)

df_mecanico
```

# EXPORTAR A EXCEL
Se exporta los maestros a un csv.

```python
df_tipo_ot.to_csv(path_output / "tipo_ot.csv", index=False, encoding='latin1')
df_frecuencia.to_csv(path_output / "frecuencia.csv", index=False, encoding='latin1')
df_tipo_taller.to_csv(path_output / "tipo_taller.csv", index=False, encoding='latin1')
df_averia.to_csv(path_output / "averia.csv", index=False, encoding='latin1')
df_taller.to_csv(path_output / "taller.csv", index=False, encoding='latin1')
df_repuesto.to_csv(path_output / "repuesto.csv", index=False, encoding='latin1')
df_mecanico.to_csv(path_output / "mecanico.csv", index=False, encoding='latin1')
```
