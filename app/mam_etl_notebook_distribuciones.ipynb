## Tabla de distribuciones

La información de las distribuciones se obtiene de la hoja "base datos 2022" del excel "Base de Datos Distribución Actual(MDI01-067) 2020.xlsx".

Se consideran unicamente dos columnas: 

* Identificador (id)
* Identificador Producto 
* Identificador Wilaya 
* Identificador Camión
* Identificador Conductor
* Fecha y hora de salida 
* Fecha y hora de llegada 
* Kilómetros de salidada
* Kilómetros de llegada 
* TM (¿Tamaño?)
* incidencias 
* Observaciones

Qué tener en cuenta: 

* Qué identificador se va a utlizar ¿n serie?
* Formato de la fecha 

```python
df_distribuciones = pd.read_excel(
    path_input / 'Base de Datos Distribución Actual (MDI01-067)  2020.xlsx',
    sheet_name='base  datos 2022',
    usecols='A, B, C, D, F, I, J, L, M, N, AC, AD',
    names=['no serie', 'conductor', 'camion','fecha_salida', 'fecha_llegada', 'km_salida', 'km_llegada',
    'tm', 'tipo_producto', 'wilaya', 'incidencias', 'observaciones'], 
    header=2
    )

df_distribuciones = df_distribuciones.iloc[0:199,] 

df_distribuciones['conductor'] = df_distribuciones['conductor'].str.title()
df_distribuciones['conductor'] = df_distribuciones['conductor'].str.lstrip()
df_distribuciones['conductor'] = df_distribuciones['conductor'].str.rstrip()

df_distribuciones['wilaya'] = df_distribuciones['wilaya'].str.capitalize()
df_distribuciones['wilaya'] = df_distribuciones['wilaya'].str.lstrip()
df_distribuciones['wilaya'] = df_distribuciones["wilaya"].str.rstrip()

# merge de la tabla distribuciones con la tabla producto 
df_distribuciones = pd.merge(left = df_distribuciones, right = df_tipo_producto, how='outer', left_on='tipo_producto', 
right_on='tipo_producto')

# merge de la tabla distribuciones con la tabla conductor
df_distribuciones["conductor"] = df_distribuciones["conductor"].str.title()
df_distribuciones = pd.merge(left = df_distribuciones, right = df_conductor, how='outer', left_on='conductor', 
right_on='conductor')

# merge de la tabla distribuciones con la tabla camion
df_distribuciones = pd.merge(left = df_distribuciones, right = df_camion, how='outer', left_on='camion', 
right_on='camion')

# merge de la tabla distribuciones con la tabla wilaya
df_distribuciones = pd.merge(left = df_distribuciones, right = df_wilaya, how='outer', left_on='wilaya', right_on='wilaya')

# se arma de nuevo el dataframe distribuciones 
d_distribuciones = {'no serie': df_distribuciones['no serie'],'id_producto': df_distribuciones['id_producto'], 'id_wilaya': df_distribuciones['id_wilaya'], 
    'id_camion': df_distribuciones['id_camion'], 'id_conductor': df_distribuciones['id_conductor'],  
    'fecha_hora_salida': df_fecha['fecha_hora_salida'], 'fecha_hora_llegada': df_fecha['fecha_hora_llegada'], 
    'km_salida': df_distribuciones['km_salida'], 'km_llegada': df_distribuciones['km_llegada'], 'tm': df_distribuciones['tm'], 
    'incidencias': df_distribuciones['incidencias'], 'observaciones': df_distribuciones['observaciones']}

df_distribuciones = pd.DataFrame(d_distribuciones).sort_values('no serie')
df_distribuciones = df_distribuciones.iloc[0:199,]
#df_distribuciones.index += 1
#df_distribuciones.reset_index(names='id_distribuciones')

df_distribuciones['no serie'] = df_distribuciones['no serie'].astype(int)
df_distribuciones['id_producto'] = df_distribuciones['id_producto'].astype(int)
df_distribuciones['id_wilaya'] = df_distribuciones['id_wilaya'].astype(int)
df_distribuciones['id_camion'] = df_distribuciones['id_camion'].astype(int)
df_distribuciones['id_conductor'] = df_distribuciones['id_conductor'].astype(int) 

df_distribuciones.head(20)
    
```
